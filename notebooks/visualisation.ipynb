```python
# üìä Visualisation √† partir de report.json
import matplotlib.pyplot as plt
import json

report_path = "../models/report.json"

# Chargement du JSON
with open(report_path, "r") as f:
    report = json.load(f)

# Extraction des donn√©es
models = list(report.keys())
params = [report[m]["Params"] for m in models]
sizes = [report[m]["Size (MB)"] for m in models]
accuracies = [report[m]["Accuracy"] for m in models]

# üîç V√©rification
print("‚úÖ Mod√®les :", models)
print("üìê Params  :", params)
print("üíæ Tailles :", sizes)
print("üéØ Accuracies :", accuracies)
```

```python
# üìà Accuracy
plt.figure(figsize=(8, 5))
plt.bar(models, accuracies, color='skyblue')
plt.title("Comparaison des mod√®les ‚Äì Accuracy")
plt.ylabel("Accuracy (%)")
plt.ylim(min(accuracies) - 0.2, max(accuracies) + 0.2)
plt.grid(axis='y')
plt.show()
```

```python
# üì¶ Taille
plt.figure(figsize=(8, 5))
plt.bar(models, sizes, color='orange')
plt.title("Comparaison des mod√®les ‚Äì Taille (MB)")
plt.ylabel("Taille du fichier .pt (MB)")
plt.grid(axis='y')
plt.show()
```

```python
# üß† Param√®tres
plt.figure(figsize=(8, 5))
plt.bar(models, params, color='green')
plt.title("Comparaison des mod√®les ‚Äì Nombre de param√®tres")
plt.ylabel("Nombre de param√®tres")
plt.grid(axis='y')
plt.show()
```
